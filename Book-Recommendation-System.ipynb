```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('data/Books.csv')
users = pd.read_csv('data/Users.csv')
ratings = pd.read_csv('data/Ratings.csv')
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_ratings_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns = {'Book-Rating': 'Num-Rating'}, inplace=True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings_df.rename(columns={'Book-Rating': 'Avg-Ratings'}, inplace=True)
avg_ratings_df
```

```python
final_df = avg_ratings_df.merge(num_ratings_df, on='Book-Title')
final_df.rename(columns={'Avg-Ratings':'Avg-Rating'}, inplace=True)
final_df
```

```python
final_df = final_df[final_df['Num-Rating'] >= 200]
final_df.sort_values('Avg-Rating', ascending=False, inplace=True)
final_df.head(50)
```

```python
popular_books_df = final_df.merge(books, on='Book-Title')
popular_books_df
```

```python
popular_books_df = popular_books_df[['Book-Title', 'Avg-Rating', 'Num-Rating', 'Book-Author']].head(50)
```

```python
popular_books_df.columns
```

```python
popular_books_df = popular_books_df.drop_duplicates(subset='Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'Num-Rating', 'Avg-Rating']]
```
