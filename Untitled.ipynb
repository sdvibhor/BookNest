```python
import pandas as pd
import numpy as np
import pickle
```

```python
books = pd.read_csv('data/Books.csv')
users = pd.read_csv('data/Users.csv')
ratings = pd.read_csv('data/Ratings.csv')
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
user_150_rating = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 150
good_readers = user_150_rating[user_150_rating].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_readers)]
```

```python
x = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = x[x].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
popular_books_df = pickle.load(open('processed-data/popular-books.pkl', 'rb'))
popular_books_df
```

```python
def recommend_books(book_name):
    book_index = np.where(pt.index == book_name)[0][0]
    book_distances = similarity_score[book_index]
    similar_books = sorted(list(enumerate(similarity_score[book_index])), key = lambda x:x[1], reverse=True)[1:11]

    data = []
    for i in similar_books:
        item = []
        temp_df = popular_books_df[popular_books_df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))


        data.append(item)

    return data
    # return suggestions
```

```python
recommend_books('Harry Potter and the Chamber of Secrets (Book 2)')
```
